## model building


```python
x=data.drop('diagnosis', axis=1)
y=data.diagnosis
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=10)
```

```python
# scaling to the data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.neural_network import MLPClassifier
model = MLPClassifier( hidden_layer_sizes=(50,10,4), #  1st layer - 50 neurons, 2nd 10 and 3rd 4 ...
                       learning_rate_init=0.1, # time to model to learn data
                       max_iter=100, # 100 times model will learn (FP+ LOSS +BP)
                       random_state=2) # randomly divide data
model.fit(X_train,y_train) ## training the data
```

```python
# PREDICTIONS
```

```python
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)
```

```python
## Evaluating the model created
from sklearn.metrics import accuracy_score,confusion_matrix
print("Train accuracy :",accuracy_score(y_train,y_pred_train))
print("Test accuracy :",accuracy_score(y_test,y_pred_test))
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score, confusion_matrix

# Generate predictions
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)

# Print accuracy
print("Train accuracy :", accuracy_score(y_train, y_pred_train))
print("Test accuracy  :", accuracy_score(y_test, y_pred_test))

# Show confusion matrix using pandas crosstab
print("Confusion Matrix:")
print(pd.crosstab(y_test, y_pred_test, rownames=['Actual'], colnames=['Predicted']))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_test))
```
